```python
import pandas as pd
import re
import nltk

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

nltk.download('punkt')
```

```python
data = pd.read_csv("support_tickets.csv")
data.head()
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.strip()
    return text

data["Cleaned_Ticket"] = data["Ticket"].apply(clean_text)
data.head()
```

```python
X = data["Cleaned_Ticket"]
y_category = data["Category"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y_category, test_size=0.2, random_state=42
)

category_model = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("classifier", MultinomialNB())
])

category_model.fit(X_train, y_train)

category_predictions = category_model.predict(X_test)

print("Category Accuracy:", accuracy_score(y_test, category_predictions))
print(classification_report(y_test, category_predictions))
```

```python
y_priority = data["Priority"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y_priority, test_size=0.2, random_state=42
)

priority_model = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("classifier", MultinomialNB())
])

priority_model.fit(X_train, y_train)

priority_predictions = priority_model.predict(X_test)

print("Priority Accuracy:", accuracy_score(y_test, priority_predictions))
print(classification_report(y_test, priority_predictions))
```

```python
def predict_ticket(ticket):
    cleaned = clean_text(ticket)
    
    category = category_model.predict([cleaned])[0]
    priority = priority_model.predict([cleaned])[0]
    
    print("Ticket:", ticket)
    print("Predicted Category:", category)
    print("Predicted Priority:", priority)

predict_ticket("My payment failed but money was deducted")
```
